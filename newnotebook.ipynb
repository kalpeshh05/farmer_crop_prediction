### ready to train the algorithm


```python
from sklearn.cluster import KMeans
```

```python
Kmeans_obj = KMeans()
```

algorithm trained and prediction

```python
labels = Kmeans_obj.fit_predict(x_transformed)  
```

```python
set(labels)
```

```python
labels
```

```python
df['cluster_no'] = labels
```
